```python
plt.figure(figsize=[12,5])
plt.subplot(1,2,1)
sns.distplot(x=df['charges'],label=df['charges'].skew().round(2))
plt.legend()
plt.subplot(1,2,2)
sns.boxplot(x=df['charges'])
```

```python
df['charges'] = np.where(df['charges']>35000,df['charges'].median(),df['charges'])
```
